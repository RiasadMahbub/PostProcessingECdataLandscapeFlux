```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import date, timedelta
```

```python
# print current working directory
cwd = os.getcwd()
print("current working directory: ", cwd)
```

```python
# set the working directory
```

```python
# Read the data
way3rawdata = pd.read_csv("Way_3_Canopy_Height_Data.csv") 
way4rawdata = pd.read_csv("Way_4_Canopy_Height_Data.csv") 
way3rawdata["date"] = pd.to_datetime(way3rawdata["Date"])
way4rawdata["date"] = pd.to_datetime(way4rawdata["Date"])
```

```python
DOPw3 = date(2022, 4, 29)
DOHw3 = date(2022, 9, 6)
DOPw4 = date(2022, 4, 28)
DOHw4 = date(2022, 9, 6)
delta = timedelta(days = 1)
```

```python
datesw3 = []
while DOPw3 <=DOHw3:
    datesw3.append(DOPw3)
    DOPw3 += delta
print(datesw3)

datesw4 = []
while DOPw4 <=DOHw4:
    datesw4.append(DOPw4)
    DOPw4 += delta
print(datesw4)
```

```python
## Convert the dates to a dataframe
datesw3_df = pd.DataFrame({'date': datesw3})
datesw4_df = pd.DataFrame({'date': datesw4})
datesw3_df
```

```python
datesw3_df['date'] = pd.to_datetime(datesw3_df['date'], utc = True)
way3rawdata['date'] = pd.to_datetime(way3rawdata['date'], utc = True)
datesw4_df['date'] = pd.to_datetime(datesw4_df['date'], utc = True)
way4rawdata['date'] = pd.to_datetime(way4rawdata['date'], utc = True)
merged_df_w3 = pd.merge(datesw3_df, way3rawdata, on = 'date', how = "left")
merged_df_w4 = pd.merge(datesw4_df, way4rawdata, on = 'date', how = "left")
pd.set_option('display.max_rows', None)

## add doy 
merged_df_w3['doy'] = merged_df_w3['date'].dt.dayofyear
merged_df_w4['doy'] = merged_df_w4['date'].dt.dayofyear
merged_df_w3
```

# Way 3 Interpolation

```python
dfway3DoyDate = merged_df_w3[['date', 'doy']].copy()
dfway3DoyDate
```

```python
merged_df_w3.S2 = merged_df_w3.Canopy_Height_Avg.interpolate()
print(merged_df_w3.S2)
```

```python
plt.plot(merged_df_w3.S2)
plt.show()
```

```python
dfway3Smerged= dfway3DoyDate.merge(merged_df_w4, how='left', on='date')
dfway3Smerged
```

```python
dfway3ToSave = dfway3Smerged[['date', 'Canopy_Height_Avg']].copy()
dfway3ToSave
```

```python
dfway3ToSave.to_csv('Dynamic_Way3.csv')
```
